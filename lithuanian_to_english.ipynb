```python
import os
import genanki
from elevenlabs.client import ElevenLabs
from dotenv import load_dotenv

# --- Configuration ---
load_dotenv()
ELEVENLABS_API_KEY = os.getenv("ELEVENLABS_API_KEY")

# Voice ID вЂ” "Rachel" is a clear, natural English voice.
# Browse all voices: client.voices.get_all()
VOICE_ID = "JBFqnCBsd6RMkjVDRZzb"

MODEL_ID = "eleven_turbo_v2_5"  # Handles accented/foreign words better than v1

AUDIO_DIR = "audio_files"
os.makedirs(AUDIO_DIR, exist_ok=True)

client = ElevenLabs(api_key=ELEVENLABS_API_KEY)

# --- Your phrases ---
facts = [
    # Greetings
    ("Labas", "Hello"),
    ("Labas rytas", "Good morning"),
    ("Labas vakaras", "Good evening"),
    ("Labanakt", "Good night"),
    ("Kaip laikaisi?", "How are you?"),
    ("AДЌiЕ«, gerai", "Thanks, I'm fine"),
    ("Iki pasimatymo", "See you later"),
    ("Iki rytojaus", "Until tomorrow"),
    ("Geros dienos", "Have a good day"),
    ("Malonu susipaЕѕinti", "Nice to meet you"),
    ("AtsipraЕЎau", "Excuse me / I'm sorry"),
    ("PraЕЎom", "You're welcome"),
    ("Sveikinu!", "Congratulations!"),
    ("SД—kmД—s!", "Good luck!"),
    ("Turiu eiti", "I have to go"),
    ("AЕЎ sutinku", "I agree"),
    ("ЕЅinoma", "Of course"),
    ("GalbЕ«t", "Maybe"),
    ("Iki!", "Bye!"),
    ("Viso gero", "Goodbye"),

    # Questions
    ("Kur esi?", "Where are you?"),
    ("KД… veiki?", "What are you doing?"),
    ("Kada grДЇЕЎi?", "When will you return?"),
    ("KodД—l vД—luoji?", "Why are you late?"),
    ("Kiek kainuoja?", "How much is it?"),
    ("Ar girdi?", "Do you hear?"),
    ("Ar Еѕinai?", "Do you know?"),
    ("Ar nori?", "Do you want?"),
    ("Ar gali?", "Can you?"),
    ("Kiek valandЕі?", "What time is it?"),
    ("Kur einam?", "Where are we going?"),
    ("Ar skauda?", "Does it hurt?"),
    ("KД… sakai?", "What are you saying?"),
    ("Ar supratai?", "Did you understand?"),
    ("Kur stotelД—?", "Where is the bus stop?"),

    # Food & Home
    ("Paduok druskД…", "Pass the salt"),
    ("Ar skanu?", "Is it tasty?"),
    ("Noriu gerti", "I want to drink"),
    ("Noriu valgyti", "I want to eat"),
    ("Esu sotus", "I am full"),
    ("Kada valgysim?", "When will we eat?"),
    ("Supjaustyk duonД…", "Slice the bread"),
    ("Atidaryk duris", "Open the door"),
    ("UЕѕdaryk langД…", "Close the window"),
    ("IЕЎjunk ЕЎviesД…", "Turn off the light"),
    ("Kur telefonas?", "Where is the phone?"),
    ("Baterija mirД—", "Battery died"),
    ("Duok kroviklДЇ", "Give me the charger"),

    # Getting Around
    ("AЕЎ pasiklydau", "I am lost"),
    ("Kur centras?", "Where is the center?"),
    ("Eik tiesiai", "Go straight"),
    ("Suk ДЇ kairД™", "Turn left"),
    ("Suk ДЇ deЕЎinД™", "Turn right"),
    ("Tai arti", "It is near"),
    ("Tai toli", "It is far"),
    ("Man reikia taksi", "I need a taxi"),
    ("Kur bilietai?", "Where are the tickets?"),
    ("Traukinys vД—luoja", "The train is late"),
    ("Kur vieЕЎbutis?", "Where is the hotel?"),
    ("Eime pД—sДЌiomis", "Let's go on foot"),
    ("Gero kelio", "Have a safe trip"),
    ("Kitas sustojimas", "Next stop"),
    ("Mes atvykome", "We arrived"),
    ("AЕЎ nesuprantu", "I don't understand"),
    ("KalbД—kite lД—ДЌiau", "Speak slower"),
    ("Pakartokite, praЕЎau", "Repeat, please"),
    ("AЕЎ uЕѕsienietis", "I am a foreigner"),
    ("PadД—kite man", "Help me"),

    # Health & Feelings
    ("AЕЎ sergu", "I am sick"),
    ("Man skauda", "It hurts me"),
    ("Man skauda galvД…", "I have a headache"),
    ("AЕЎ pavargau", "I got tired"),
    ("Noriu miego", "I want to sleep"),
    ("AЕЎ laimingas", "I am happy"),
    ("NeliЕ«dД—k", "Don't be sad"),
    ("Nebijok", "Don't be afraid"),
    ("Viskas gerai", "Everything is fine"),
    ("Bus geriau", "It will be better"),
    ("Man bloga", "I feel sick (nauseous)"),
    ("Gerk vandenДЇ", "Drink water"),
    ("Reikia daktaro", "Need a doctor"),
    ("Man ЕЎalta", "I am cold"),
    ("Man karЕЎta", "I am hot"),
    ("Gerk ЕЎiltД… arbatД…", "Drink warm tea"),
    ("Reikia vaistЕі", "Need medicine"),
    ("AЕЎ geriau jauДЌiuosi", "I feel better"),
    ("Viskas bus gerai", "Everything will be fine"),
    ("Niekada nepasiduok", "Never give up"),
    ("TikД—k savimi", "Believe in yourself"),
    ("Myliu tave", "I love you"),
    ("AДЌiЕ« uЕѕ viskД…", "Thanks for everything"),

    # Shopping
    ("Noriu pirkti", "I want to buy"),
    ("Kokia kaina?", "What is the price?"),
    ("Per brangu", "Too expensive"),
    ("Tai pigu", "It's cheap"),
    ("Duok nuolaidД…", "Give a discount"),
    ("MokД—siu kortele", "I'll pay by card"),
    ("MokД—siu grynais", "I'll pay cash"),
    ("KvitД…, praЕЎau", "The receipt, please"),
    ("Per maЕѕas", "Too small"),
    ("Per didelis", "Too big"),
    ("Reikia bankomato", "Need an ATM"),
    ("SД…skaitД…, praЕЎau", "The bill, please"),

    # Daily Life
    ("Palauk minutД™", "Wait a minute"),
    ("Daryk dabar", "Do it now"),
    ("PradД—kim", "Let's start"),
    ("Baigta!", "Finished!"),
    ("Kas naujo?", "What's new?"),
    ("AЕЎ pamirЕЎau", "I forgot"),
    ("AЕЎ prisimenu", "I remember"),
    ("RuoЕЎkis", "Get ready"),
    ("VД—luojame!", "We are late!"),
    ("Skubam!", "Let's hurry!"),

    # Social
    ("Seniai matyti", "Long time no see"),
    ("UЕѕsuk ДЇ sveДЌius", "Come over for a visit"),
    ("AЕЎ paskambinsiu", "I will call"),
    ("ParaЕЎyk ЕѕinutД™", "Write a message"),
    ("Noriu pasikalbД—ti", "I want to talk"),
    ("Niekam nesakyk", "Don't tell anyone"),
    ("Tu esi teisus", "You are right"),
    ("AЕЎ klydau", "I was wrong"),
    ("Atleisk man", "Forgive me"),
    ("BЕ«k atsargus", "Be careful"),
    ("Tai juokinga", "That is funny"),
    ("AЕЎ juokauju", "I am joking"),
    ("Klausyk atidЕѕiai", "Listen carefully"),
    ("Ar supranti mane?", "Do you understand me?"),
    ("PaaiЕЎkink dar kartД…", "Explain once more"),
    ("Kuo tu vardu?", "What is your name?"),
    ("Kaip sekasi?", "How is it going?"),
    ("KД… tu veiki?", "What are you doing?"),
    ("Kur yra tualetas?", "Where is the toilet?"),
    ("Galiu tau padД—ti?", "Can I help you?"),
    ("AДЌiЕ« uЕѕ pagalbД…", "Thanks for the help"),
    ("PraЕЎau prisД—sti", "Please sit down"),
    ("AЕЎ tave myliu", "I love you"),
    ("Susitiksime rytoj", "We will meet tomorrow"),
]


# --- ElevenLabs TTS ---
def generate_audio_elevenlabs(text: str, output_path: str) -> bool:
    """Generate audio via ElevenLabs SDK and save to output_path. Returns True on success."""
    try:
        audio = client.text_to_speech.convert(
            text=text,
            voice_id=VOICE_ID,
            model_id=MODEL_ID,
            output_format="mp3_44100_128",
            voice_settings={"stability": 0.5, "similarity_boost": 0.75},
        )
        with open(output_path, "wb") as f:
            for chunk in audio:
                f.write(chunk)
        return True
    except Exception as e:
        print(f"  ElevenLabs error for '{text}': {e}")
        return False


# --- Anki Model ---
my_model = genanki.Model(
    1607392321,
    'Lithuanian -> English + Audio',
    fields=[
        {'name': 'Lithuanian'},
        {'name': 'English'},
        {'name': 'Audio'},
    ],
    templates=[
        {
            'name': 'Card 1',
            'qfmt': '{{Lithuanian}}',
            'afmt': '{{Lithuanian}}<hr id="answer">{{English}}<br><br>{{Audio}}',
        },
    ]
)

my_deck = genanki.Deck(2059400114, 'Lithuanian -> English')
media_files = []
cards_created = 0

for lithuanian, english in facts:
    safe_name = "".join(
        c for c in english.replace(" ", "_") if c.isalnum() or c == "_"
    )[:40]
    audio_filename = f"{safe_name}.mp3"
    audio_full_path = os.path.join(AUDIO_DIR, audio_filename)

    # Generate audio only if not already cached
    if not os.path.exists(audio_full_path):
        print(f"Generating: {english}")
        success = generate_audio_elevenlabs(english, audio_full_path)
        if not success:
            continue
    else:
        print(f"Cached:     {english}")

    note = genanki.Note(
        model=my_model,
        fields=[lithuanian, english, f"[sound:{audio_filename}]"],
        tags=['lithuanian', 'english'],
    )
    my_deck.add_note(note)
    media_files.append(audio_full_path)
    cards_created += 1

# --- Save Package ---
package = genanki.Package(my_deck)
package.media_files = media_files
package.write_to_file('lithuanian_to_english.apkg')

print(f"\nDone! {cards_created} cards saved to lithuanian_to_english.apkg")
```
