```python
import os
import genanki
from gtts import gTTS

# --- Configuration ---
AUDIO_DIR = "audio_files"
os.makedirs(AUDIO_DIR, exist_ok=True)

# Correct codes for British English
TTS_LANG = 'en'
TTS_TLD  = 'co.uk'  # This is the correct TLD for a British accent

# --- Anki Model ---
my_model = genanki.Model(
    1607392321,
    'Lithuanian -> English + Audio',
    fields=[
        {'name': 'Lithuanian'},
        {'name': 'English'},
        {'name': 'Audio'},
    ],
    templates=[
        {
            'name': 'Card 1',
            'qfmt': '{{Lithuanian}}',
            # Added {{Audio}} to back side
            'afmt': '{{Lithuanian}}<hr id="answer">{{English}}<br><br>{{Audio}}',
        },
    ]
)

my_deck = genanki.Deck(2059400114, 'Lithuanian -> English')
media_files = []

for lithuanian, english in facts:
    # 1. Create a safe filename
    safe_name = "".join([c for c in english.replace(" ", "_") if c.isalnum() or c == '_'])[:40]
    audio_filename = f"{safe_name}.mp3"
    audio_full_path = os.path.join(AUDIO_DIR, audio_filename)

    # 2. Generate Audio if it doesn't exist
    if not os.path.exists(audio_full_path):
        try:
            tts = gTTS(text=english, lang=TTS_LANG, tld=TTS_TLD, slow=False)
            tts.save(audio_full_path)
        except Exception as e:
            print(f"Failed for {english}: {e}")
            continue

    # 3. Create Anki Note
    # IMPORTANT: The tag in the field must just be [sound:filename.mp3]
    sound_tag = f"[sound:{audio_filename}]"
    
    note = genanki.Note(
        model=my_model,
        fields=[lithuanian, english, sound_tag],
        tags=['lithuanian', 'english']
    )
    my_deck.add_note(note)
    
    # 4. Add the ABSOLUTE path to the media list
    media_files.append(audio_full_path)

# --- Save Package ---
package = genanki.Package(my_deck)
package.media_files = media_files # Assign the list of paths
package.write_to_file('lithuanian_to_british.apkg')

print(f"Success! {len(facts)} cards created.")
```
